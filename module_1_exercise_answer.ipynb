## 3\. Strings

Aplique três métodos distintos na *string* abaixo, você pode conferir alguns métodos neste [link](https://www.w3schools.com/python/python_ref_string.asp):

```
cancao = 'Roda mundo, roda gigante, roda moinho, roda pião.'
print(cancao.upper())
print(cancao.lower())
print(cancao.replace('roda', 'wheel'))
```

Extraia da string abaixo o valor da taxa **selic** na variável `selic` e o valor do **ano** na variavel `ano`. Imprima os valores na tela.

```
noticia = 'Selic vai a 2,75% e supera expectativas; é a primeira alta em 6 anos.'
posicao_selic = noticia.find('%')
posicao_ano = noticia.find('ano')
selic = noticia[posicao_selic - 4:posicao_selic]
ano = noticia[posicao_ano - 2:posicao_ano]
print(selic)
print(ano)
```

## 4\. Booleanos

Utilize a tabela da verdade para responder: qual o valor da variável x?

```
a = False
b = True

x = not a & b
```

Como não há a separação por parenteses entre o **not** e a operação lógica **a & b**, primeiro será executado o **not a** e o resultado disso será comparado com **b**. Sendo assim, temos a resposta como **True** conforme abaixo:

**not a** = True

**b** = True

**True & True** = True

ou

x = not False & True

x = True & True

x = True

Conforme a tabela verdade, o resultado do operador lógico **AND** é **True** somente quando todas as variáveis, ou operações lógicas, retornam **True**.
